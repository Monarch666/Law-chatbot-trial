```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from docx import Document
import re
```

```python
class IndianLawyerChatbot:
```

```python
  def __init__(self, model_name="gpt2"):
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModelForCausalLM.from_pretrained(model_name)
        self.context = ""
```

```python
def import_docx(self, "D:\t_h_e\Business Model\THE.docx"):
        try:
            doc = Document("D:\t_h_e\Business Model\THE.docx")
            full_text = []
            for para in doc.paragraphs:
                full_text.append(para.text)
            self.context += "\n".join(full_text)
            print(f"Successfully imported document: {"D:\t_h_e\Business Model\THE.docx"}")
        except Exception as e:
            print(f"Error importing document: {str(e)}")

    # ... (other methods remain the same)
```

```python
 def chat(self):
        print("Indian Lawyer Chatbot: Hello! I'm here to assist you with legal queries. How can I help you today?")
        print("To import a document, type 'import' followed by the file path, e.g., 'import C:\\Documents\\legal_case.docx'")
        
        while True:
            user_input = input("You: ")
            if user_input.lower() in ['exit', 'quit', 'bye']:
                print("Indian Lawyer Chatbot: Thank you for using our service. Goodbye!")
                break
            
            if user_input.lower().startswith("import "):
                file_path = user_input[7:].strip()
                self.import_docx(file_path)
                continue
            
            response = self.generate_response(user_input)
            print(f"Indian Lawyer Chatbot: {response}")
```

```python

if __name__ == "__main__":
    chatbot = IndianLawyerChatbot()
    chatbot.chat()
```

```python
from flask import Flask, request, jsonify
import os

app = Flask(__name__)

# Initialize the chatbot
chatbot = IndianLawyerChatbot()

@app.route('/chat', methods=['POST'])
def chat():
    data = request.json
    user_input = data.get('message')
    
    if user_input.lower().startswith("import "):
        file_path = user_input[7:].strip()
        chatbot.import_docx(file_path)
        return jsonify({"response": f"Imported document: {file_path}"})
    
    response = chatbot.generate_response(user_input)
    return jsonify({"response": response})

if __name__ == "__main__":
    app.run(debug=True)
```
